```python
import pandas as pd
from sklearn.model_selection import train_test_split
import sklearn.impute as skl
import numpy as np
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

**Preprocessing dataset**:

```python
#code
dt1 = pd.read_csv("preprocessing_weather.csv")
simputer = skl.SimpleImputer(missing_values = np.nan, strategy='median', fill_value=None,
verbose=0, copy=True, add_indicator=False)
dt1[['temperature','humidity']] = simputer.fit_transform(dt1[['temperature','humidity']])
print(dt1)
```

# Task 2.1. dataset1.csv


```python
#code
data1 = pd.read_csv('dataset1.csv')
X_train = data1[['x']]
Y_train = data1[['y']]
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
print(regressor.intercept_)
print(regressor.coef_)
y_pred = regressor.predict(X_train)
print(y_pred)
metrics.r2_score(Y_train,y_pred)
```

# Task 2.2. dataset2.csv


```python
#code
data2 = pd.read_csv('dataset2.csv')
x_train, x_test, y_train, y_test = train_test_split(data2[['x1','x2','x3']],data2['y'], test_size=0.3, train_size=0.7,
random_state=None, shuffle=True, stratify=None)
# print(x_train)
# print(x_test)
# print(y_train)
# print(y_test)
regressor = LinearRegression()
regressor.fit(x_train,y_train)
print(regressor.intercept_)
print(regressor.coef_)
y_pred = regressor.predict(x_train)
print(y_pred)
metrics.r2_score(y_train,y_pred)
```
